```python
import pandas as pd
import numpy as np
from sklearn import cross_validation

data_csv = pd.read_csv('menkyo_info.csv',header=None)
data_csv.columns = ['moji','label']
moji = data_csv['moji']
#moji_array = np.array(moji)
moji_list = moji.tolist()

label = data_csv['label']
#label_array = np.array(label)
#print(label_array)
lbl_list = label.tolist()
label_li = []
for lbl in lbl_list:
    for i in range(5):
        label_li.append(lbl)
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences


moji_str =  ''
for mojis in moji_list:
    moji_str += mojis

tokenizer = Tokenizer()
tokenizer.fit_on_texts(moji_str)
moji_seq = tokenizer.texts_to_sequences(moji_str)
vocab_size = len(tokenizer.word_index) + 1
```

```python


    
    
import numpy as np 


point = np.load('zahyou.npy')
#print(point)
xmin = []
ymin = []
xmax = []
ymax = []
for p in point:
    xmin.append(p[0])
    ymin.append(p[1])
    xmax.append(p[2])
    ymax.append(p[3])
    
#print(len(xmin_zahyou))

# 文字と座標は独立していて良い！！それぞれ一つ一つにラベルが振ってあれば問題なし
#data=[文字列、座標]
data = []
#moji_zahyou = []
for idx in range(len(xmin)):
    
    data.append(moji_seq[idx])
    
    data.append([xmin[idx]])

    data.append([ymin[idx]])
    
    data.append([xmax[idx]])
    
    data.append([ymax[idx]])
    
    #data.append(moji_zahyou)
#print('data',len(data))
#print('xmins',len(xmins))
#print(data)


count_max = []
#moji_zahyou = []
for idx in range(len(xmin)):
    
    count_max.append(moji_seq[idx][0])
    
    count_max.append(xmin[idx])

    count_max.append(ymin[idx])
    
    count_max.append(xmax[idx])
    
    count_max.append(ymax[idx])
    
max_num = max(count_max)

print(data)
```

```python
from keras.layers import Dense,LSTM, Activation
from keras.models import Sequential
from keras.utils import np_utils
from keras.layers.embeddings import Embedding
from keras.layers.convolutional import Conv1D, MaxPooling1D
from keras.optimizers import Adadelta,Adam


#moji_train = np.array(moji_train)
#moji_test = np.array(moji_test)
#label_train = np.array(label_train)
#label_test = np.array(label_test)


labels = list(set(label_li))
label_dic = {}
for idx, lbl in enumerate(labels):
     label_dic[idx] = lbl

print(label_dic)

nb_label = len(labels)
#print(nb_label)
#print(label_train)

label_list =[]
for label in label_li:
    for key , val in label_dic.items():
        if val == label:
            label_list.append(key)
        else:continue
        
    
moji_train, moji_test, label_train, label_test = cross_validation.train_test_split(data, label_list)


moji_train = np.array(moji_train)
moji_test = np.array(moji_test)
label_train2 = np.array(label_train)
label_test2 = np.array(label_test)

print('label_train',label_train)

for train in moji_train:
    num = len(train)


#reshape(データセットの数＝LSTMのタイムステップ、一度に入力させたいデータの数＝データを何個使って予測をさせるか、データ１種類につき１個のデータだから１)
#moji_tra = moji_train.reshape( len(moji_train) , 1, 1)
moji_tra = moji_train.reshape( len(moji_train) , num)
#moji_tes = moji_test.reshape( len(moji_test) , 1, 1)
moji_tes = moji_test.reshape( len(moji_test) , num)
#label_tra = label_train.reshape( len(label_train) , 1, 1)
#label_tes = label_test.reshape( len(label_test) , 1, 1)
label_train = np_utils.to_categorical(label_train2, nb_label)
label_test = np_utils.to_categorical(label_test2, nb_label)




batch_size = 2

# 出力する時の次元数。CNNと違って中間層は入力データによって変わるから。
# 何種類のデータを一度に出力するか
neurons = 70
neurons2 = 70
neurons3 = 100


#This is the size of the vocabulary in the text data. 
#For example, if your data is integer encoded to values between 0-10, then the size of the vocabulary would be 11 words.
#  入力の形や出力の形には影響しない
# 単語をtexts_to_sequences() で整数にするんだが、その整数の最大値+1を入力する
#  tokenizer = Tokenizer()
#vocab_size = len(tokenizer.word_index) + 1 これで単語を整数にした時の最大の整数＋１がわかる。
# 単語をベクトルにした時の最大の数字をかく

#[[16 12 18 17 16]
# [ 8 14  6 15 19]]なら１９＋１で２０
#num_word = vocab_size 
num_word = max_num +1

# ここから二つが出力の形に関わる。出力はinput_length が行数になり、embedding_vecor_lengthが列数次元数になる


# 出力した時の一単語につき〜次元で出力　埋め込みベクトルのサイズを示す。単語の数でいいと思われ
#[[16 12 18 17 16]
# [ 8 14  6 15 19]]  なら、ここを１０に設定すると、この5列が10列に引き延ばされる。ただの数字の整数を何次元まで伸ばすか＝何列まで引き延ばすか
#１０にすると、「16」→[ 0.028 0.021 0.005 -0.024 0.024 -0.037 -0.039 -0.013 -0.035 0. ]
# 特に決まりはなさげ。前に学習した単語が１００次元ならここも１００にしたりとか４次元のベクトルに変換した後に８次元にしたりとか
embedding_vecor_length = 2

#input_length argumetは各入力シーケンスのサイズを決定します。入力するリストの中のリストのデータの個数
#  それぞれ入力する時の文字数、リストの中のリストの中の要素の数！！！
#[[16 12 18 17 16]
# [ 8 14  6 15 19]]  ならinput_length = 5
# この数が出力時の配列の行数になる！！一つ一つの単語をembedding_vecor_length次元まで伸ばして、一要素ずつ（リストの中のリストの中の要素数）縦に置いていくイメージ。
# 一要素ずつ縦に置いていく時にここで指定した数の分だけ、縦に配置される
#[[2 3 1 2 1]
# [3 3 0 2 3]]が入力で　input_length =５なら
#[[[ 0.03530972 -0.01738765]
#  [-0.01431809  0.02838255]
#  [-0.00195207  0.04445304]
#  [ 0.03530972 -0.01738765]
#  [-0.00195207  0.04445304]]
#
# [[-0.01431809  0.02838255]
#  [-0.01431809  0.02838255]
#  [-0.00543046 -0.01861401]
#  [ 0.03530972 -0.01738765]
#  [-0.01431809  0.02838255]]]
#　それぞれのリストの中のリストの中の数字が縦に置かれて、それが横へ１次元から２次元へ引き伸ばされている
#  入力の列の数が出力の行数になる！！
input_length = 1

def build_model(train_data , neurons ,neurons2,neurons3, batich_size):
    model = Sequential()
    #model.add(Embedding(num_words, embedding_vecor_length, input_length=max_speech_length))
    model.add(Embedding(num_word,embedding_vecor_length,input_length = input_length))
    model.add(Conv1D(32, 3, border_mode='same', activation='relu'))
    #model.add(LSTM(neurons, batch_input_shape= (batch_size, train_data.shape[1], train_data.shape[2]) ))
    #  ,stateful=True
    model.add(LSTM(neurons, batch_input_shape= (batch_size, train_data.shape[1]), activation='relu',return_sequences = True ))
    model.add(LSTM(neurons2,return_sequences = True ))
    model.add(LSTM(neurons3,return_sequences = False ))
    #model.add(Dense(100, activation='relu'))
    model.add(Dense(7, activation='softmax'))
    #Adadelta(lr=1.0, rho=0.95, epsilon=None, decay=0.0)
    Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0)
    model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
    return model


#  fitは学習データは3次元のarrayで渡す,ラベルは１次元のarrayで渡す！！
#  作ったモデルを関数にして、modelインスタンスに格納してfitで使わないと動かない 

model = build_model(moji_tra , neurons , neurons2 , neurons3 ,1)
```

```python
model.fit(moji_tra, label_train, epochs=30 ,batch_size=batch_size,validation_data=(moji_tes, label_test))
model.reset_states()
```

```python
predict = model.predict(moji_tes,batch_size=batch_size)
print(predict)
model.reset_states()
```

```python
#  改善案
predicted = model.predict(moji_seq , batch_size=batch_size)
print(predicted)
model.reset_states()
```

```python
# 座標も含めてどれに該当するか
answer1 = []
for pre in predict:
    for idx , num in enumerate(pre):
        if num == np.max(pre):
            ans = label_dic[idx]
            answer1.append(ans)

label = []
for lbl_test in label_test2:
    lbl = label_dic[lbl_test]
    label.append(lbl)
            
print('\n正解',' 　　　　', '予想', sep='\t')   



for idx in range(len(label_test)):
    
    print(label[idx],'　　',answer1[idx],sep='\t')
    
```
